```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
godfather=open("C:\\Users\sycha\\Desktop\\files\\2020\\winter\\YBIGTA\\git\\14_wordcloud_konlp_0204\\data\\thegodfather.txt", "r", encoding="UTF8").read()
godfather
```

```python
stopwords
```

```python
godfatherwc = WordCloud(background_color="white",
                       collocations=True,
                       stopwords=stopwords)
godfatherwc.generate(godfather)
```

```python
godfatherwc.words_
```

```python
godfatherwc.to_image()
```

```python
godfatherwc = WordCloud(background_color="white",
                       collocations=True,
                       stopwords=stopwords,
                       width=800, height=450)
godfatherwc.generate(godfather)
godfatherwc.to_image()
```

```python
stopwords.update(['s', 'said', 'one', 'Il'])
godfatherwc.generate(godfather)
godfatherwc.to_image()
```

```python
godfatherwc = WordCloud(background_color="black",
                       collocations=True,
                       stopwords=stopwords,
                       width=800, height=450,
                       colormap='flag')
godfatherwc.generate(godfather)
godfatherwc.to_image()
```

```python
godfatherwc = WordCloud(background_color="black",
                       collocations=True,
                       stopwords=stopwords,
                       width=800, height=450,
                       colormap='flag',
                       min_font_size=5)
godfatherwc.generate(godfather)
godfatherwc.to_image()
```
