## 2. 단축키

Google Colaboratory에서는 Jupyter Notebook에서 사용하시던 단축키 앞에 __'Ctrl + M'__을 추가해주시면 대부분이 동일하게 작동합니다!
※ 저의 개인적인 생각으로는 Colab에서는 Jupyter와 달리 초록색, 파란색으로 셀의 상태를 구분하지 않아서 Ctrl + M을 추가해준 것 같습니다.

- **Ctrl + M + A** : 바로 위에 새로운 Cell 생성

- **Ctrl + M + B** : 바로 밑에 새로운 Cell 생성

- **Ctrl + M + D** : 해당 셀을 삭제합니다.

- **Ctrl + M + Z** : 삭제한 셀을 다시 생성

- **Ctrl + M + Y** : 셀을 Python Code 셀로 지정

- **Ctrl + M + M** : 셀을 글 작성을 위한 Markdown 셀로 지정

- **Shift + Enter** : 해당 셀 실행
